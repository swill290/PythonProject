# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = "SELECT ConvertedComp FROM master"
df = pd.read_sql_query(QUERY, conn)
print(df.head())  # Check the first few rows to confirm the data

import matplotlib.pyplot as plt

# Plot a histogram of 'ConvertedComp'
plt.figure(figsize=(10, 6))  # Set the size of the plot
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='blue', edgecolor='black')  # Remove NaN values
plt.title('Histogram of ConvertedComp')  # Set the title of the plot
plt.xlabel('ConvertedComp')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.show()  # Display the plot
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = "SELECT Age FROM master"
df_age = pd.read_sql_query(QUERY, conn)
print(df_age.head())  # Check the first few rows to confirm the data

import seaborn as sns

# Plot a box plot of 'Age'
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_age, x='Age', color='green')
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

import matplotlib.pyplot as plt

# Query the data
QUERY = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(QUERY, conn)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='purple')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Query the data
QUERY = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(QUERY, conn)

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5, color='orange')  # Scale the bubble size by Age
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Age as bubble size)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import pandas as pd

# Query to list all tables
QUERY = """
SELECT name as Table_Name
FROM sqlite_master
WHERE type = 'table'
"""
# Run the query and view the tables
tables_df = pd.read_sql_query(QUERY, conn)
print(tables_df)

QUERY = "PRAGMA table_info(DatabaseDesireNextYear)"  # Replace with the relevant table name
columns_df = pd.read_sql_query(QUERY, conn)
print(columns_df)

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
top_databases_df = pd.read_sql_query(QUERY, conn)
print(top_databases_df)

import matplotlib.pyplot as plt

# Assuming `top_databases_df` contains 'DatabaseDesireNextYear' and 'count' columns
plt.figure(figsize=(8, 8))
plt.pie(
    top_databases_df['count'],
    labels=top_databases_df['DatabaseDesireNextYear'],
    autopct='%1.1f%%'
)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Query the data
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_age_30_35 = pd.read_sql_query(QUERY, conn)

# Calculate the median values
median_work_week_hrs = df_age_30_35['WorkWeekHrs'].median()
median_code_rev_hrs = df_age_30_35['CodeRevHrs'].median()

# Create the stacked bar chart
plt.bar('30-35', median_work_week_hrs, label='WorkWeekHrs')
plt.bar('30-35', median_code_rev_hrs, bottom=median_work_week_hrs, label='CodeRevHrs')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Query the data
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_age_45_60 = pd.read_sql_query(QUERY, conn)

# Group by age and calculate the median ConvertedComp
df_grouped = df_age_45_60.groupby('Age')['ConvertedComp'].median()

# Plot the line chart
plt.figure(figsize=(10, 6))
df_grouped.plot()
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation by Age (45-60)')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

# Query the data
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""
df_main_branch = pd.read_sql_query(QUERY, conn)

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
df_main_branch.plot(kind='barh', x='MainBranch', y='Count')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branch')
plt.show()
```
